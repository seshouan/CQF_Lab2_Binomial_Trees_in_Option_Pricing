```python
# Ignore Warnings
import warnings
warnings.filterwarnings('ignore')
# Import Pandas, Numpy and Scipy
import pandas as pd
import numpy as np
from scipy.stats import norm
# Import plotly express
import plotly.express as px
import plotly.graph_objects as go
px.defaults.width, px.defaults.height = 1000, 600
```

```python
# read in the dataframe
df = pd.read_csv('CQF_Jan_24_M1L1_Completed_Excel.csv', index_col=0, parse_dates=True)['2011':'2013']
df.tail()
```

```python
# Visualize the plot to verify the data
px.line(df, y="Adj Close", title="SPY Historical Chart")
```

```python
# Verify the datetime format
df.index
```

```python
# Get last 300 index values
df.index[-300:]
```

```python
# Calculate returns and add it to existing DataFrame as a column
df['Return'] = df['Adj Close'].pct_change().fillna(0)
df.head()
```

```python
# Calculate mean and sigma of returns
mu = np.mean(df['Return'])
sigma = np.std(df['Return'])

# Calculate the scaled return : zscore
df['Scaled Return'] = df['Return'].apply(lambda x: (x-mu)/sigma)

# Check the output
df.head()
```

```python
 # Generate some data (replace with your scaled returns data)
data = df['Scaled Return']

# Create the normal distribution plot
x = np.linspace(data.min(), data.max(), 200)
y = norm.pdf(x, data.mean(), data.std())      # same as -> y = 1/np.sqrt(2*np.pi)*np.exp(-(x**2)/2)

# Create the histogram
fig = px.histogram(data, x=data, histnorm='probability density', nbins=200, labels={'x':''}, title='Empirical vs Normal distributions')

# Add the Norm distribution curve
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Normal Distribution'))

# Show the graph
fig.show()
```
